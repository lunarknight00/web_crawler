```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# data sourcing from url below
# https://finance.yahoo.com/quote/MSFT/history?p=MSFT
url = "MSFT.csv"
data = pd.read_csv(url,parse_dates=["Date"])
```

```python
data
```

```python
data.plot(x="Date",y="Close")
```

```python
import bokeh.plotting as bp
bp.output_file("images/msft_1.html",title = "Bokeh Example(static)")
p = bp.figure(title = "Historical Stock Quotes",x_axis_type = "datetime")
p.line(data["Date"],data["Close"],legend = "MSFT")
bp.show(p)
```
